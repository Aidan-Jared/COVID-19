```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/covid_19_data.csv')
```

```python
df.head()
```

```python
df["ObservationDate"] = df['ObservationDate'].apply(pd.to_datetime)
df.drop(["SNo"], axis = 1, inplace = True)
df.head()
```

```python
countries = df["Country/Region"].unique().tolist()
print("{} countries have been affected by the virst".format(len(countries)))
print(countries)
```

```python
cases = df.groupby("Country/Region")["Confirmed"].sum()
cases["Country/Region"] = cases.index
cases = cases.reset_index()
cases
```

```python
df.groupby(["Country/Region", "Province/State"]).sum()
```

```python
deaths = df.groupby(["Country/Region"])["Deaths"].sum()
deaths
```

```python
rec = df.groupby(["Country/Region"])["Recovered"].sum()
rec
```

```python
plot_df = df[["Country/Region","Confirmed", "Deaths" , "Recovered"]].groupby("Country/Region").sum()
plot_df.sort_values(["Confirmed"], ascending=False).head(10).plot.bar()
plt.show()
```

```python
plot_df.sort_values(["Deaths"], ascending=False).head(10).plot.bar()
plt.show()
```

```python
plot_df.sort_values(["Recovered"], ascending=False).head(10).plot.bar()
plt.show()
```

```python
def dataPloter(df, Country):
    df[df["Country/Region"] == Country].groupby("ObservationDate").sum().plot.line()
    plt.show()
```

```python
dataPloter(df, "Mainland China")
```

```python
dataPloter(df, "Italy")
```

```python
dataPloter(df, "South Korea")
```

```python
dataPloter(df, "US")
```
